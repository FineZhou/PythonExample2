# 饼状图

饼状图显示一个数据序列中各项大小占各项总和的比例，
饼状图的数据点显示为整个饼状图的百分比，
如前十大品牌占市场份额图

```python
labels=["A","B","C","D"]
fracs=[15,30,45,10]
plt.pie(x=fracs,labels=labels)
plt.show()
#这样画出的饼状图呈椭圆形，原因是x与y的比例不是1:1，如果要显示成正圆，就需要在绘制前加上plt.axes(aspect=1)
```

```python
plt.axes(aspect=1)
plt.pie(x=fracs,labels=labels)
plt.show()
```

```python
#加上具体的比例值
plt.axes(aspect=1)
plt.pie(x=fracs,labels=labels,autopct="%.0f%%")
plt.show()
```

```python
#突出显示某项比例
plt.axes(aspect=1)
explode=[0,0.05,0.08,0]#对第二第三个labels突出显示了，值越大越偏离中心
plt.pie(x=fracs,labels=labels,autopct="%.0f%%",explode=explode)
plt.show()
```

```python
#加上阴影,属性：shadow=True
plt.axes(aspect=1)
explode=[0,0.05,0.08,0]
plt.pie(x=fracs,labels=labels,autopct="%.0f%%",explode=explode,shadow=True)
plt.show()
```

```python
labels=["SH","BJ","SZ","GD"]
fracs=[20,10,30,35]
explode=[0,0,0.04,0] #SZ突出显示
plt.axes(aspect=1) #饼状图呈圆型，
plt.pie(x=fracs,labels=labels,autopct="%.00f%%",explode=explode,shadow=True)
plt.show()
```

# 箱型图

箱形图又叫盒须图，盒式图或箱线图，
是一种用作显示一组数据分散情况资料的统计图，
上边缘、上四份位数，中位数，下四分位数，下边缘，异常值.

它能显示出一组数据的最大值、最小值、中位数、及上下四份位数

```python
np.random.seed(100)
data=np.random.normal(size=1000,loc=0,scale=1)
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data,sym="*")#异常值部分显示成*
plt.show()
```

```python
#whis默认是1.5
plt.boxplot(data,sym="o",whis=0.05)#whis用来展示异常值部分长短，越大，异常值越小，反之越大
plt.show()
```

```python
#同时画多个箱形图
data=np.random.normal(size=(1000,4),loc=0,scale=1)
labels=["A","B","C","D"]
plt.boxplot(data,labels=labels)
plt.show()
```

```python
salary=[2710,3000,2850,3500,4200,2755,2880,2890,2920,2940,3325,3130]
labels=["Accountant"]
plt.boxplot(salary,labels=labels)
plt.show()
```

# 面向对象形式绘图

```python
x=np.arange(1,10)
```

```python
y=np.random.randn(len(x))
```

*****创建子图****

```python
fig=plt.figure()#创建画布
#创建x轴 返回Axes实例 参数111的三个参数“1”意思是：
#第一个参数1：子图总行数，
#第二个参数1：子图总列数
#第三个参数1：子图的位置
ax=fig.add_subplot(111)  
l,=plt.plot(x,y)
t=ax.set_title("object oriented")
```

```python
fig=plt.figure()
ax=fig.add_subplot(121)
ax.plot(x,y)
ax2=fig.add_subplot(122)
ax2.plot(x,-y)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(221)
ax.plot(x,y)
ax.set_title("image 1")
ax2=fig.add_subplot(222)
ax2.plot(x,-y)
ax2.set_title("image 2")
ax3=fig.add_subplot(223)
ax3.plot(x,x**2)
ax3.set_title("image 3")
ax4=fig.add_subplot(224)
ax4.plot(x,x**x)
ax4.set_title("image 4")
plt.show()
```

```python
#展示网格
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y)
ax.set_title("image 1")
#ax.grid(True)
ax.grid(linestyle="-.",color="r",linewidth=2)
plt.show()
```

```python
#展示图例
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,label="Normal")
ax.plot(x,y*2,label="Fast")
ax.plot(x,y*3,label="Faster")
ax.set_title("image 1")
ax.grid(True)
#ax.legend() #默认是loc=0，ncol=1,ncol表示一行显示几个图例图标
ax.legend(loc=0,ncol=2)  #loc为（0,1,2,3,4）中的一个值，1为右上角，0和2为左上角，3为左下角，4为右下角
plt.show()
```
